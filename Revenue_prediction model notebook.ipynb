```python
X2 = pd.read_csv(r"C:\Users\Lenovo\Desktop\movies_analysis\X_final.csv")
X2
```

```python
print(X2.dtypes)
```

```python
# Fill null values with 0
X2.fillna(0, inplace=True)
X2
```

```python
#our equation is in the form of y = f(x)
# y = revenue
#seperation of y and x as independent elements of the equation
y = df1['revenue']
# Fill null values with 0
y.fillna(0, inplace=True)
y
```

```python
y1 = pd.read_csv(r"C:\Users\Lenovo\Desktop\movies_analysis\y.csv")
y1
```

```python
#Data splitting
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X2,y1, test_size = 0.2, random_state = 100)
```

```python
X_train
```

```python
X_test
```

```python
#Model Building
#Linear Regression
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

#train the model
lr.fit(X_train, y_train)
```

```python
#Applying the model to make a prediction using train data
y_lr_train_pred = lr.predict(X_train)
y_lr_train_pred
```

```python
y_lr_test_pred = lr.predict(X_test)
y_lr_test_pred
```

```python
#Evaluate Model Perfomance
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test,y_lr_test_pred )
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
print('LR MSE (TRAIN):', lr_train_mse)
print('LR R2 (TRAIN):', lr_train_r2)

print('LR MSE (TEST):', lr_test_mse)
print('LR R2 (TEST):', lr_test_r2)
```

```python
#Random forest
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the model on the training data
rf_model.fit(X_train, y_train)
```

```python
# Make predictions on the train data
rf_y_train_pred = rf_model.predict(X_train)
rf_y_train_pred
```

```python
# Make predictions on the test data
rf_y_test_pred = rf_model.predict(X_test)
rf_y_test_pred
```

```python
#Evaluate Model Perfomance

rf_train_mse = mean_squared_error(y_train, rf_y_train_pred)
rf_train_r2 = r2_score(y_train, rf_y_train_pred)

rf_test_mse = mean_squared_error(y_test, rf_y_test_pred)
rf_test_r2 = r2_score(y_test, rf_y_test_pred)
```

```python
print('RF MSE (TRAIN):', rf_train_mse)
print('RF R2 (TRAIN):', rf_train_r2)

print('RF MSE (TEST):', rf_test_mse)
print('RF R2 (TEST):', rf_test_r2)
```

```python
rf_results = pd.DataFrame(['Random Forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
rf_results
```

```python
lr_results = pd.DataFrame(['Linear Regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
lr_results
```

```python
#Combining the 2 tables
df_models = pd.concat([lr_results, rf_results], axis = 0)
df_models
```
